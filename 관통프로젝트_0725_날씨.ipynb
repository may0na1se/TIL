```python
import requests

# 0 데이터 받기
def request_weather_data(cityname = 'seoul', my_api_key = "867916a1fd161bcd7eae7b771e3cc2e9"):
    # 요청시 헤더정보를 크롬으로 지정
    request_headers = {
    'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
    AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
    Safari/537.36'), }

    url = f'https://api.openweathermap.org/data/2.5/weather?q={cityname}&appid={my_api_key}'
    data = requests.get(url, headers=request_headers).json()
    return data



# 1 key값 출력
def print_key():
    data = request_weather_data()
    return data




# 2 특정 데이터 추출
def extract_target_data():
    data = request_weather_data()
    export_data = {
        'main' : data['main'],
        'weather': data['weather']
    }
    return export_data


# 3 데이터 한글 변환
def translate_data():
    data = request_weather_data()
    kor_data = {
        "좌표": {
            "경도": data['coord']['lon'],
            "위도": data['coord']['lat']
        },
        "날씨": {
            "날씨 ID": data['weather'][0]['id'],
            "주요 날씨": data['weather'][0]['main'],
            "날씨 설명": data['weather'][0]['description'],
            "아이콘 코드": data['weather'][0]['icon']
        },
        "기반 정보": data['base'],
        "주요 기상 정보": {
            "현재 기온": data['main']['temp'],
            "체감 기온": data['main']['feels_like'],
            "최저 기온": data['main']['temp_min'],
            "최고 기온": data['main']['temp_max'],
            "기압": data['main']['pressure'],
            "습도": data['main']['humidity'],
            "해수면 기압": data['main']['sea_level'],
            "지면 기압": data['main']['grnd_level']
        },
        "가시거리": data['visibility'],
        "바람 정보": {
            "풍속": data['wind']['speed'],
            "풍향": data['wind']['deg']
        },
        "구름 정보": {
            "구름량": data['clouds']['all']
        },
        "데이터 시간": data['dt'],
        "시스템 정보": {
            "시스템 타입": data['sys']['type'],
            "시스템 ID": data['sys']['id'],
            "국가": data['sys']['country'],
            "일출 시각": data['sys']['sunrise'],
            "일몰 시각": data['sys']['sunset']
        },
        "시간대": data['timezone'],
        "도시 ID": data['id'],
        "도시 이름": data['name'],
        "응답 코드": data['cod']
    }
    return kor_data


# 4 한글 데이터에 섭씨 온도를 추가
def add_celsius():
    data = translate_data()
    for i in ["현재 기온", "체감 기온", "최저 기온", "최고 기온"]:
        k = data['주요 기상 정보'][i]
        c = k - 273.15
        data['주요 기상 정보'].update({f'(섭씨) {i}' : round(c,2)})
    return data

# 5 AI 활용하기
```
